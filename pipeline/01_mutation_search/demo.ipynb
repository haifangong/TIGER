# Step 1 Demo — Mutational Search & Activity Pre-filter

This notebook reproduces a **small** mutational search:
1. choose a short template peptide
2. enumerate `k`-point mutants with physicochemical filters
3. classify activity with the bundled CatBoost model
4. inspect positive / negative CSV outputs

> Tip: start with `k=1` for a fast demo. Production runs often use `k=2` or `k=3`.


````python
from pathlib import Path
import pandas as pd

# Ensure we run relative to this step folder
STEP_DIR = Path('.').resolve()
print('Working directory:', STEP_DIR)
assert (STEP_DIR / 'search_mutations.py').exists(), 'Please open/run this notebook from 01_mutation_search/'
````

````python
# Configuration for a fast interactive demo
TEMPLATE = 'KSMLKSMK'   # short template used in POAP examples
K = 1                   # mutation depth (increase carefully)
OUT_DIR = STEP_DIR / 'outputs' / 'notebook_demo'

print(f'Template={TEMPLATE}  k={K}  out={OUT_DIR}')
````

````python
from search_mutations import run_search

pos_csv, neg_csv = run_search(
    sequence=TEMPLATE,
    search_length=K,
    output_dir=OUT_DIR,
)
print('Positive CSV:', pos_csv)
print('Negative CSV:', neg_csv)
````

````python
# Inspect outputs
pos = pd.read_csv(pos_csv, header=None, names=['seq', 'seq_copy']) if pos_csv.stat().st_size else pd.DataFrame(columns=['seq','seq_copy'])
neg = pd.read_csv(neg_csv, header=None, names=['idx', 'seq']) if neg_csv.stat().st_size else pd.DataFrame(columns=['idx','seq'])
print(f'positives={len(pos)}  negatives={len(neg)}')
display(pos.head())
display(neg.head())
````
